```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import albumentations as A
import random 
transform = A.Compose([
    A.OneOf([
        A.HorizontalFlip(),
        A.VerticalFlip(),
    ], p=0.9),
    A.Rotate(limit=360, p=0.5),
])
```

```python
data = pd.read_csv('datasets/training_data.csv')
data.head()
```

```python
len(data)
```

```python
data_1 = data[data['class']==2]
len(data_1)
```

```python
data_new = data.copy()
for i in range(5):
    for i in range(len(data)):
        if data['class'].loc[i] == 1:
            image = plt.imread(f"images/{data['image_id'].loc[i]}.jpg")
            image = transform(image = image)['image']
            result_str = ''.join((random.choice('abcdxyzpqr') for i in range(5)))
            image_name = f"{data['image_id'].loc[i]}_{result_str}"
            print(image_name)
            plt.imsave(f"images/{image_name}.jpg", image)
            new_row = {
                'Unnamed: 0': data['Unnamed: 0'].iloc[i],  
                'healthy': 0, 
                'multiple_diseases': 1,
                'rust': 0,
                'scab': 0,
                'class': 1,
                'image_id': image_name
            }
         
            data_new = pd.concat([data_new, pd.DataFrame([new_row])], ignore_index=True)

```
